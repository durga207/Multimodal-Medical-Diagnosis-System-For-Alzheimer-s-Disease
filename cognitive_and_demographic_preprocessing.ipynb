```python
dataset_root = './dataset/'
```

```python
import os
import pandas as pd

ad_folder = dataset_root + 'preprocessed_dataset/AD_dataset'
cn_folder = dataset_root + 'preprocessed_dataset/CN_dataset'
csv_file = dataset_root + 'ADNIMERGE.csv'

adni_merge_df = pd.read_csv(csv_file)

ad_patient_ids = [pid for pid in os.listdir(ad_folder) if os.path.isdir(os.path.join(ad_folder, pid))]
cn_patient_ids = [pid for pid in os.listdir(cn_folder) if os.path.isdir(os.path.join(cn_folder, pid))]

selected_patient_ids = ad_patient_ids + cn_patient_ids

adni_merge_filtered = adni_merge_df[adni_merge_df['PTID'].isin(selected_patient_ids)]

selected_columns = ['PTID', 'VISCODE', 'ADAS13', 'FDG', 'TAU', 'PTAU', 'CDRSB', 'MMSE',
                    'RAVLT_immediate', 'RAVLT_learning', 'RAVLT_forgetting', 'RAVLT_perc_forgetting',
                    'FAQ', 'MOCA', 'Hippocampus', 'AGE', 'PTGENDER', 'PTEDUCAT', 'APOE4']

adni_merge_filtered = adni_merge_filtered[adni_merge_filtered['VISCODE'].isin(['bl', 'm12', 'm24', 'm36', 'm48'])]

adni_merge_filtered['LABEL'] = 0 
adni_merge_filtered.loc[adni_merge_filtered['PTID'].isin(ad_patient_ids), 'LABEL'] = 1 

adni_merge_filtered = adni_merge_filtered[['PTID', 'VISCODE', 'ADAS13', 'FDG', 'TAU', 'PTAU', 'CDRSB', 'MMSE',
                                            'RAVLT_immediate', 'RAVLT_learning', 'RAVLT_forgetting', 'RAVLT_perc_forgetting',
                                            'FAQ', 'MOCA', 'Hippocampus', 'AGE', 'PTGENDER', 'PTEDUCAT', 'APOE4', 'LABEL']]

adni_merge_filtered
```

```python
missing_viscode_patients = []
for patient_id in selected_patient_ids:
    if not set(['bl', 'm12', 'm24', 'm36', 'm48']).issubset(set(adni_merge_filtered[adni_merge_filtered['PTID'] == patient_id]['VISCODE'])):
        missing_viscode_patients.append(patient_id)

if missing_viscode_patients:
    print("Patients with missing VISCODEs:", missing_viscode_patients)
else:
    print("All patients have VISCODEs bl, m12, m24, m36 and m48")
```

```python
output_csv_path = dataset_root + 'adni_merge_filtered.csv'

adni_merge_filtered.to_csv(output_csv_path, index=False)
```

```python
import numpy as np

processed_adni_merge = adni_merge_filtered.copy()

processed_adni_merge['PTGENDER'] = processed_adni_merge['PTGENDER'].replace({'Male': 0, 'Female': 1})

processed_adni_merge['VISCODE'] = processed_adni_merge['VISCODE'].replace({'bl': 0, 'm12': 1, 'm24': 2, 'm36': 3, 'm48': 4})
processed_adni_merge['VISCODE'] = processed_adni_merge['VISCODE'].astype(int)
processed_adni_merge['TAU'] = pd.to_numeric(processed_adni_merge['TAU'], errors='coerce')
processed_adni_merge['PTAU'] = pd.to_numeric(processed_adni_merge['PTAU'], errors='coerce')

processed_adni_merge['TAU'] = processed_adni_merge['TAU'].astype(float)
processed_adni_merge['PTAU'] = processed_adni_merge['PTAU'].astype(float)
```

```python
processed_adni_merge = processed_adni_merge.apply(lambda col: col.fillna(col.mean()) if col.name != 'PTID' else col)
```

```python
output_csv_path = 'processed_adni_merge.csv'

processed_adni_merge.to_csv(output_csv_path, index=False)
```

```python
processed_adni_merge_values_only = processed_adni_merge.drop(columns=['PTID'])

for column in processed_adni_merge_values_only.columns:
    max_value = processed_adni_merge_values_only[column].max()
    min_value = processed_adni_merge_values_only[column].min()
    mean_value = processed_adni_merge_values_only[column].mean()
    print(f"Column: {column}, Max: {max_value}, Min: {min_value}, , Mean: {mean_value}")
```

```python
from sklearn.preprocessing import MinMaxScaler

exclude_columns = ['PTID', 'VISCODE', 'FDG', 'AGE', 'PTGENDER', 'PTEDUCAT', 'LABEL']

scaler = MinMaxScaler()
processed_adni_merge_normalized = processed_adni_merge
processed_adni_merge_normalized = processed_adni_merge_normalized.drop(columns=exclude_columns)
processed_adni_merge_normalized = pd.DataFrame(scaler.fit_transform(processed_adni_merge_normalized), columns=processed_adni_merge_normalized.columns)
print(processed_adni_merge_normalized)

for column in exclude_columns:
    processed_adni_merge_normalized[column] = processed_adni_merge[column].values

for column in exclude_columns:
    print(processed_adni_merge[column])

processed_adni_merge_normalized = processed_adni_merge_normalized[['PTID', 'VISCODE'] + [col for col in processed_adni_merge_normalized.columns if col not in ['PTID', 'VISCODE']]]
print(processed_adni_merge_normalized)
```

```python
output_csv_path = 'processed_adni_merge_normalized.csv'

processed_adni_merge_normalized.to_csv(output_csv_path, index=False)
```
